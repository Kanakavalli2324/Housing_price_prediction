```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
from sklearn.model_selection import train_test_split
x=data.drop( ['median_house_value'], axis=1 )
y=data['median_house_value']
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2)
train_data=x_tr.join(y_tr)
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data['total_rooms']= np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']= np.log(train_data['total_bedrooms']+1)
train_data['population']= np.log(train_data['population']+1)
train_data['households']= np.log(train_data['households']+1)
train_data.hist(figsize=(15,8))
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity,dtype=int))
train_data=train_data.drop(['ocean_proximity'],axis=1)
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude",y="longitude",data=train_data,hue="median_house_value",palette="coolwarm")
```

```python
train_data['b_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['h_ratio']=train_data['total_rooms']/train_data['households']
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression

x_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']

rgr=LinearRegression()
rgr.fit(x_train,y_train)
```

```python
test_data=x_te.join(y_te)
test_data['total_rooms']= np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']= np.log(test_data['total_bedrooms']+1)
test_data['population']= np.log(test_data['population']+1)
test_data['households']= np.log(test_data['households']+1)
```

```python
test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity,dtype=int))
```

```python
test_data=test_data.drop(['ocean_proximity'],axis=1)
```

```python
test_data['b_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['h_ratio']=test_data['total_rooms']/test_data['households']
```

```python
from sklearn.ensemble import RandomForestRegressor

reg=RandomForestRegressor()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
reg=RandomForestRegressor()

param={
    "n_estimators":[100,200,300],
    "min_samples_split": [2,4],
    "max_depth": [None,4,8]
}
g_s=GridSearchCV(reg,param,cv=5,scoring="neg_mean_squared_error",return_train_score=True)
g_s.fit(x_train,y_train)
```

```python
b_f=g_s.best_estimator_
```

```python
b_f.score(x_test,y_test)
```
